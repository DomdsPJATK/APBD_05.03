Wgraj zawartość pliku orders.csv do ramki danych.
Przyjrzyj się danym, wykorzystując poznane metody (describe, info, head) zanalizuj dane.

Następnie:



1. Zamień order_date na format daty. Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień;
2. Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;
3. Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;
4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.
5. Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016.


```python
import pandas as pd
import numpy as np

data = pd.read_csv('orders.csv', sep=",")
orders = pd.DataFrame(data)
orders
```

```python
# 1

orders['order_date'] = pd.to_datetime(orders['order_date'], format='%Y-%m-%d')

orders
```

```python
# 2

orders2 = orders.groupby('tshirt_category').agg(count=('tshirt_category', 'count')).reset_index()

orders2
```

```python
# 3

def normalize(x):
  x = x.replace('Wh ', 'White ').replace('Bl ', 'Black ').replace('Tshirt', 'T-Shirt')
  return x;

orders['tshirt_category'] = orders['tshirt_category'].apply(normalize)

orders
```

```python
# 4

orders[['tshirt_colour', 'tshirt_type', 'tshirt_size']] = orders['tshirt_category'].apply(lambda x: pd.Series([np.nan, x, np.nan] if 'T-Shirt' not in x else x.split(' ')))

orders
```

```python
# 5

orders[(orders['order_date'] >= '2014-12-31') &
       (orders['order_date'] <= '2016-08-02')].sort_values(by='order_date', ascending=True)
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv('customers.csv', sep=",")
customers = pd.DataFrame(data)

customers
```

```python
# 1

for col in customers.columns:
    print(col)
```

```python
# 2

print(f'Pierwsza tabela miała {orders.shape[0]} wierszy i {orders.shape[1]} column.\n'
      f'Druga tabela ma {customers.shape[0]} wierszy i {customers.shape[1]} column.')
```

```python
# 3

customers.rename(columns={'customerID': 'customer_id'}, inplace=True)

customers
```

```python
# 4
# Join wykorzystuje metode merge w srodku, używa indexów do złączenia tabel, można wykorzystać łączenie po kluczu, jest wydaniejsze od concat (Informacja z dokumentacji)
# Użycie join pozwoliło mi dodać index który był potrzebny w nastepnym zadaniu

merged = orders.set_index('customer_id').join(customers.set_index('customer_id'))

merged
```
